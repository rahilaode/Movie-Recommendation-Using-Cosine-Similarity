# Features Selection

In this project i will select feature mostly used to find similarities. I will use select genres of movies, keywords, taglinge, cast or artist in the movie, and the director.

```python
features_selection = ['genres', 'keywords', 'tagline', 'cast', 'director']
features_selection
```

See again the null value in the selected features

```python
df[features_selection].isnull().sum()
```

We can see that the selected features have null values. So i will fill the null values with empty string.

```python
for feature in features_selection:
    df[feature] = df[feature].fillna('')

df[features_selection].isnull().sum()
```

After that, i will combine all selected features into one string for each index.  

```python
features_combination = df[features_selection[0]]+' '+df[features_selection[1]]+' '+df[features_selection[2]]+' '+df[features_selection[3]]+' '+df[features_selection[4]]
print('Data type : {}'.format(type(features_combination)))
print(features_combination)
```

```python
features_combination[0]
```

# Convert a collection of raw documents to a matrix of TF-IDF features

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(features_combination)
print(feature_vector)
```

Values above represents:

(__Index__, __token_id__), and the value following the tuple represents the __tf-idf score__ of a given token in a given document

We can see token id with __vectorizer.vocabulary___ methods

```python
print(vectorizer.vocabulary_)
```

```python
feature_vector.shape
```

Feature vector shape represents:

(__index count, token_id count__)

# Getting similarity scores

In this section i will make a matrix contains similarity score for each movies.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
similarity.shape
```

# Get suggested movie from input movie name

Get input movie from user

```python
input_movie = input('Movie Name : ')
```

Get the list of movie name from the dataset

```python
list_movie = df['title'].tolist()
print(list_movie)
```

Next, i will find the close match movie in dataset from input movie using difflib

```python
import difflib

find_close_match = difflib.get_close_matches(input_movie, list_movie)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

Next, i will find the index from close match movie

```python
movie_index = df[df['title'] == close_match]['index'].values[0]
print(movie_index)
```

Try to find similarity movie index based similarity score

```python
similarity_score = list(enumerate(similarity[movie_index]))
similarity_score
```

```python
len(similarity_score)
```

```python
sorted_similar_movie = sorted(similarity_score, key = lambda x:x[1], reverse = True)
sorted_similar_movie
```

Find the similarity movie name

```python
print("Movie search result : {}\n".format(input_movie))

i=1
for movie in sorted_similar_movie:
    index = movie[0]
    title = df[df.index == index]['title'].values[0]
    if(i<=20):
        print('{}. {}'.format(i, title))
        i+=1
```
